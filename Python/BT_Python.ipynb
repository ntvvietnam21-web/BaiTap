```python
#Bài 14
import random

def tao_mang_ngau_nhien(m, n):
    matrix = []
    for i in range(m):
        dong = []
        for j in range(n):
            dong.append(random.randint(1, 100))
        matrix.append(dong)
    return matrix

def in_mang(matrix):
    print("Bảng số hiện tại:")
    for dong in matrix:
        for phan_tu in dong:
            print(f"{phan_tu:<5}", end="")
        print()

# b & c: Xuất dòng/cột k
def xuat_dong_cot(matrix, k):
    # Xuất dòng
    if 0 <= k < len(matrix):
        print(f"Dòng {k} là:")
        for value in matrix[k]:
            print(f"{value:<5}", end="")
        print()
    else:
        print("Chỉ số dòng không hợp lệ!")

    # Xuất cột (hiển thị đẹp hơn)
    if 0 <= k < len(matrix[0]):
        print(f"Cột {k} là:")
        for i in range(len(matrix)):
            print(f"{matrix[i][k]:<5}")
    else:
        print("Chỉ số cột không hợp lệ!")

# d: Dòng có tổng lớn nhất
def tim_dong_tong_max(matrix):
    max_tong = -1
    dong_chi_so = 0
    for i in range(len(matrix)):
        tong_dong_nay = sum(matrix[i])
        if tong_dong_nay > max_tong:
            max_tong = tong_dong_nay
            dong_chi_so = i
    print(f"Dòng {dong_chi_so} có tổng lớn nhất là: {max_tong}")

# e: Cột có tích nhỏ nhất
def tim_cot_tich_min(matrix):
    so_dong = len(matrix)
    so_cot = len(matrix[0])
    min_tich = 999999999
    cot_chi_so = 0

    for j in range(so_cot):
        tich_cot_nay = 1
        for i in range(so_dong):
            tich_cot_nay *= matrix[i][j]
        if tich_cot_nay < min_tich:
            min_tich = tich_cot_nay
            cot_chi_so = j
    print(f"Cột {cot_chi_so} có tích nhỏ nhất là: {min_tich}")

# f: Dòng chẵn (0, 2, 4) và Cột lẻ (1, 3, 5)
def xuat_dong_chan_cot_le(matrix):
    print("Các số ở vị trí (Dòng chẵn, Cột lẻ):")
    for i in range(len(matrix)):
        for j in range(len(matrix[0])):
            if i % 2 == 0 and j % 2 != 0:
                print(f"Tại [{i}][{j}] có số: {matrix[i][j]}")

# g: Trung bình cộng số CHẴN ở dòng LẺ
def tbc_so_chan_dong_le(matrix):
    tong = 0
    dem = 0
    for i in range(len(matrix)):
        if i % 2 != 0:
            for so in matrix[i]:
                if so % 2 == 0:
                    tong += so
                    dem += 1
    if dem > 0:
        print(f"TBC số chẵn ở dòng lẻ: {tong / dem}")
    else:
        print("Không tìm thấy số chẵn nào ở dòng lẻ.")

# h: Trung bình cộng các số nằm ở VIỀN (biên)
def tbc_bien(matrix):
    so_dong = len(matrix)
    so_cot = len(matrix[0])
    tong = 0
    dem = 0
    for i in range(so_dong):
        for j in range(so_cot):
            if i == 0 or i == so_dong - 1 or j == 0 or j == so_cot - 1:
                tong += matrix[i][j]
                dem += 1
    print(f"TBC các số nằm ở biên: {tong / dem}")
    # i:	Tính trung bình tích các phần tử không thuộc biên.
def tbc_tich_noi_bo(matrix):
    so_dong = len(matrix)
    so_cot = len(matrix[0])

    tich = 1
    dem = 0

    for i in range(1, so_dong - 1):
        for j in range(1, so_cot - 1):
            tich *= matrix[i][j]
            dem += 1

    if dem > 0:
        print(f"Trung bình tích các phần tử không thuộc biên: {tich / dem}")
    else:
        print("Không có phần tử nào không thuộc biên.")

# --- CHẠY THỬ ---
m = int(input("Nhập số dòng: "))
n = int(input("Nhập số cột: "))

matrix_a = tao_mang_ngau_nhien(m, n)
in_mang(matrix_a)
for row_idx in range(m):
  for col_idx in range(n):
    print(row_idx, col_idx)

print("\n--- KẾT QUẢ ---")
xuat_dong_cot(matrix_a, 0)
tim_dong_tong_max(matrix_a)
tim_cot_tich_min(matrix_a)
xuat_dong_chan_cot_le(matrix_a)
tbc_so_chan_dong_le(matrix_a)
tbc_bien(matrix_a)
tbc_tich_noi_bo(matrix_a)
```

```python
#Bài 15
import datetime
# ===== a. Cấu trúc SINHVIEN =====
class SinhVien:
    def __init__(self, ma_sv, ten, nam_sinh, diem_tb):
        self.ma_sv = ma_sv
        self.ten = ten
        self.nam_sinh = nam_sinh
        self.diem_tb = diem_tb
    def __str__(self):
        return f"{self.ma_sv:<12}{self.ten:<25}{self.nam_sinh:<10}{self.diem_tb:<10}"

# ===== In danh sách =====
def in_danh_sach(ds):
    print("\nDANH SÁCH SINH VIÊN")
    print(f"{'Mã SV':<12}{'Tên':<25}{'Năm sinh':<10}{'Điểm TB':<10}")
    for sv in ds:
        print(sv)

# ===== b. Nhập danh sách =====
def nhap_danh_sach():
    ds = []
    n = int(input("Nhập số sinh viên: "))
    for i in range(n):
        print(f"\nNhập sinh viên thứ {i+1}")
        while True:
            ma = input("Mã SV (10 ký tự): ")
            if len(ma) == 10:
                break
            print("Mã phải đủ 10 ký tự!")
        ten = input("Tên: ")
        nam = int(input("Năm sinh: "))
        diem = float(input("Điểm TB: "))
        ds.append(SinhVien(ma, ten, nam, diem))
    return ds

# ===== c. Đếm SV lên lớp =====
def dem_len_lop(ds):
    dem = sum(1 for sv in ds if sv.diem_tb >= 5)
    print(f"Số SV đủ điều kiện lên lớp: {dem}")

# ===== d. SV đủ 20 tuổi =====
def sv_20_tuoi(ds):
    nam_ht = datetime.datetime.now().year
    print("\nSinh viên đủ 20 tuổi:")
    for sv in ds:
        if nam_ht - sv.nam_sinh >= 20:
            print(sv)

# ===== e. SV hệ DH =====
def dem_he_dh(ds):
    dem = sum(1 for sv in ds if len(sv.ma_sv) >= 4 and sv.ma_sv[2:4] == "DH")
    print(f"Số SV hệ đại học: {dem}")

# ===== f. SV tên Lan =====
def dem_ten_lan(ds):
    dem = 0
    for sv in ds:
        ten_cuoi = sv.ten.strip().split()[-1].lower()
        if ten_cuoi == "lan":
            dem += 1
    print(f"Số SV tên Lan: {dem}")

# ===== g. SV họ Phan =====
def dem_ho_phan(ds):
    dem = 0
    for sv in ds:
        ho = sv.ten.strip().split()[0].lower()
        if ho == "phan":
            dem += 1
    print(f"Số SV họ Phan: {dem}")

def menu():
    print("\n========= MENU =========")
    print("1. Nhập danh sách sinh viên")
    print("2. Xuất danh sách")
    print("3. Đếm SV lên lớp")
    print("4. Xuất SV đủ 20 tuổi")
    print("5. Đếm SV hệ đại học")
    print("6. Đếm SV tên Lan")
    print("7. Đếm SV họ Phan")
    print("0. Thoát")
# ===== Main =====
ds_sv = []
while True:
    menu()
    chon = input("Chọn chức năng: ")
    if chon == "1":
        ds_sv = nhap_danh_sach()
    elif chon == "2":
        if ds_sv:
            in_danh_sach(ds_sv)
        else:
            print("Chưa có dữ liệu!")
    elif chon == "3":
        dem_len_lop(ds_sv)
    elif chon == "4":
        sv_20_tuoi(ds_sv)
    elif chon == "5":
        dem_he_dh(ds_sv)
    elif chon == "6":
        dem_ten_lan(ds_sv)
    elif chon == "7":
        dem_ho_phan(ds_sv)
    elif chon == "0":
        print("Thoát chương trình.")
        break
    else:
        print("Lựa chọn không hợp lệ!")
```

```python
# Bài 2 BTVN
# ===== Nhập ma trận =====
def nhap_ma_tran():
    m = int(input("Nhập số dòng: "))
    n = int(input("Nhập số cột: "))

    a = []
    for i in range(m):
        dong = []
        for j in range(n):
            x = int(input(f"a[{i}][{j}] = "))
            dong.append(x)
        a.append(dong)
    return a

# ===== In ma trận =====
def in_ma_tran(a):
    print("\nMa trận:")
    for dong in a:
        for x in dong:
            print(f"{x:<5}", end="")
        print()

# ===== a. Tổng tam giác trên của đường chéo phụ =====
def tong_tam_giac_tren_cheo_phu(a):
    n = len(a)
    tong = 0
    for i in range(n):
        for j in range(n):
            if i + j <= n - 1:
                tong += a[i][j]
    print(f"Tổng tam giác trên của đường chéo phụ: {tong}")

# ===== b. Đổi số âm thành trị tuyệt đối =====
def doi_am_thanh_duong(a):
    for i in range(len(a)):
        for j in range(len(a[0])):
            if a[i][j] < 0:
                a[i][j] = abs(a[i][j])

# ===== c. Thay số chẵn bằng x =====
def thay_chan_bang_x(a, x):
    for i in range(len(a)):
        for j in range(len(a[0])):
            if a[i][j] % 2 == 0:
                a[i][j] = x

# ===== d. Kiểm tra toàn chẵn =====
def kiem_tra_toan_chan(a):
    for dong in a:
        for x in dong:
            if x % 2 != 0:
                print("Ma trận KHÔNG toàn số chẵn")
                return
    print("Ma trận TOÀN số chẵn")

# ===== e. Kiểm tra ma trận đối xứng =====
def kiem_tra_doi_xung(a):
    n = len(a)
    for i in range(n):
        for j in range(n):
            if a[i][j] != a[j][i]:
                print("Ma trận KHÔNG đối xứng")
                return
    print("Ma trận ĐỐI XỨNG")
a = nhap_ma_tran()
in_ma_tran(a)
print("\n--- KẾT QUẢ ---")

# a
if len(a) == len(a[0]):
    tong_tam_giac_tren_cheo_phu(a)
else:
    print("Không phải ma trận vuông nên không tính câu a")

# b
doi_am_thanh_duong(a)
print("\nSau khi đổi số âm thành trị tuyệt đối:")
in_ma_tran(a)

# c
x = int(input("\nNhập x để thay số chẵn: "))
thay_chan_bang_x(a, x)
print("Sau khi thay số chẵn:")
in_ma_tran(a)

# d
kiem_tra_toan_chan(a)

# e
if len(a) == len(a[0]):
    kiem_tra_doi_xung(a)
else:
    print("Không phải ma trận vuông nên không kiểm tra đối xứng")
```
