```python
# Code mẫu

import os
import math

def GetWinner(board):
    """
    Trả về người chiến thắng trên bàn cờ hiện tại nếu có,
    ngược lại trả về None.
    """
    # Kiểm tra hàng ngang (Horizontal)
    if board[0] == board[1] and board[1] == board[2]:
        return board[0]
    elif board[3] == board[4] and board[4] == board[5]:
        return board[3]
    elif board[6] == board[7] and board[7] == board[8]:
        return board[6]

    # Kiểm tra hàng dọc (Vertical)
    elif board[0] == board[3] and board[3] == board[6]:
        return board[0]
    elif board[1] == board[4] and board[4] == board[7]:
        return board[1]
    elif board[2] == board[5] and board[5] == board[8]:
        return board[2]

    # Kiểm tra đường chéo (Diagonal)
    elif board[0] == board[4] and board[4] == board[8]:
        return board[0]
    elif board[2] == board[4] and board[4] == board[6]:
        return board[2]

    return None

def PrintBoard(board):
    """
    Xóa màn hình console và in bàn cờ hiện tại.
    """
    os.system('cls' if os.name == 'nt' else 'clear')
    print(f'''
 {board[0]} | {board[1]} | {board[2]}
 {board[3]} | {board[4]} | {board[5]}
 {board[6]} | {board[7]} | {board[8]}
    ''')

def GetAvailableCells(board):
    """
    Trả về danh sách các chỉ số (indices) của các ô còn trống.
    """
    available = list()
    for cell in board:
        if cell != "X" and cell != "O":
            available.append(cell)
    return available

def minimax(position, depth, alpha, beta, isMaximizing):
    """
    Thuật toán AI chịu trách nhiệm chọn nước đi tốt nhất.
    Sử dụng cắt tỉa Alpha-Beta và tính toán theo độ sâu để thắng nhanh nhất.
    """
    winner = GetWinner(position)
    if winner != None:
        # Nếu X thắng trả về 10 - độ sâu, nếu O thắng trả về -10 + độ sâu
        return 10 - depth if winner == "X" else -10 + depth

    if len(GetAvailableCells(position)) == 0:
        return 0

    if isMaximizing:
        maxEval = -math.inf
        for cell in GetAvailableCells(position):
            position[cell - 1] = "X"
            eval = minimax(position, depth + 1, alpha, beta, False)
            position[cell - 1] = cell
            maxEval = max(maxEval, eval)
            alpha = max(alpha, eval)
            if beta <= alpha:
                break # Cắt tỉa (Prune)
        return maxEval
    else:
        minEval = +math.inf
        for cell in GetAvailableCells(position):
            position[cell - 1] = "O"
            eval = minimax(position, depth + 1, alpha, beta, True)
            position[cell - 1] = cell
            minEval = min(minEval, eval)
            beta = min(beta, eval)
            if beta <= alpha:
                break # Cắt tỉa (Prune)
        return minEval

def FindBestMove(currentPosition, AI):
    """
    Tìm kiếm nước đi tối ưu nhất cho AI bằng cách duyệt qua tất cả ô trống.
    """
    bestVal = -math.inf if AI == "X" else +math.inf
    bestMove = -1

    for cell in GetAvailableCells(currentPosition):
        currentPosition[cell - 1] = AI
        # Nếu AI là X thì lượt tiếp theo là lượt của người chơi (Minimize)
        isNextMaximizing = False if AI == "X" else True
        moveVal = minimax(currentPosition, 0, -math.inf, +math.inf, isNextMaximizing)
        currentPosition[cell - 1] = cell

        if AI == "X":
            if moveVal > bestVal:
                bestMove = cell
                bestVal = moveVal
        elif AI == "O":
            if moveVal < bestVal:
                bestMove = cell
                bestVal = moveVal

    return bestMove

def main():
    player = input("Play as X or O? ").strip().upper()
    AI = "O" if player == "X" else "X"
    currentGame = [*range(1, 10)] # Khởi tạo bàn cờ [1, 2, ..., 9]

    # X luôn đi trước
    currentTurn = "X"
    counter = 0

    while True:
        if currentTurn == AI:
            # AI thực hiện nước đi
            cell = FindBestMove(currentGame, AI)
            currentGame[cell - 1] = AI
            currentTurn = player
        elif currentTurn == player:
            # Người chơi thực hiện nước đi
            PrintBoard(currentGame)
            while True:
                try:
                    humanInput = int(input("Enter Number: ").strip())
                    if humanInput in GetAvailableCells(currentGame):
                        currentGame[humanInput - 1] = player
                        currentTurn = AI
                        break
                    else:
                        PrintBoard(currentGame)
                        print("Cell Not Available.")
                except ValueError:
                    print("Please enter a valid number.")

        # Kiểm tra kết thúc trò chơi
        winner = GetWinner(currentGame)
        if winner != None:
            PrintBoard(currentGame)
            print(f"{winner} WON!!!")
            break

        counter += 1
        if GetWinner(currentGame) == None and len(GetAvailableCells(currentGame)) == 0:
            PrintBoard(currentGame)
            print("Tie.")
            break

if __name__ == "__main__":
    main()
```

```python
#Code

import os
import math

# Các hằng số để code rõ ràng hơn
X, O = "X", "O"

def in_ban_co(ban_co):
    """Hiển thị bàn cờ một cách đẹp mắt"""
    os.system('cls' if os.name == 'nt' else 'clear')
    print("\n")
    for i in range(0, 9, 3):
        print(f" {ban_co[i]} | {ban_co[i+1]} | {ban_co[i+2]} ")
        if i < 6: print("---+---+---")
    print("\n")

def kiem_tra_thang(b):
    """Kiểm tra tất cả các trường hợp thắng: hàng, cột, chéo"""
    khung_thang = [
        (0,1,2), (3,4,5), (6,7,8), # Hàng ngang
        (0,3,6), (1,4,7), (2,5,8), # Hàng dọc
        (0,4,8), (2,4,6)           # Đường chéo
    ]
    for i, j, k in khung_thang:
        if b[i] == b[j] == b[k]:
            return b[i]
    return None

def lay_o_trong(ban_co):
    """Trả về danh sách các số (ô chưa bị đánh)"""
    return [o for o in ban_co if isinstance(o, int)]

def minimax(ban_co, do_sau, alpha, beta, la_nguoi_max):
    """Thuật toán Minimax kết hợp cắt tỉa Alpha-Beta"""
    nguoi_thang = kiem_tra_thang(ban_co)

    # Trạng thái dừng (Base case)
    if nguoi_thang == X: return 10 - do_sau
    if nguoi_thang == O: return -10 + do_sau
    if not lay_o_trong(ban_co): return 0

    if la_nguoi_max:
        max_eval = -math.inf
        for o in lay_o_trong(ban_co):
            ban_co[o-1] = X
            eval = minimax(ban_co, do_sau + 1, alpha, beta, False)
            ban_co[o-1] = o # Hoàn tác (Backtrack)
            max_eval = max(max_eval, eval)
            alpha = max(alpha, eval)
            if beta <= alpha: break
        return max_eval
    else:
        min_eval = math.inf
        for o in lay_o_trong(ban_co):
            ban_co[o-1] = O
            eval = minimax(ban_co, do_sau + 1, alpha, beta, True)
            ban_co[o-1] = o # Hoàn tác (Backtrack)
            min_eval = min(min_eval, eval)
            beta = min(beta, eval)
            if beta <= alpha: break
        return min_eval

def tim_nuoc_di_tot_nhat(ban_co, quan_ai):
    """AI tìm ô có điểm số cao nhất (nếu là X) hoặc thấp nhất (nếu là O)"""
    tot_nhat = -math.inf if quan_ai == X else math.inf
    o_chon = None
    la_max = (quan_ai == X)

    for o in lay_o_trong(ban_co):
        ban_co[o-1] = quan_ai
        diem = minimax(ban_co, 0, -math.inf, math.inf, not la_max)
        ban_co[o-1] = o

        if (la_max and diem > tot_nhat) or (not la_max and diem < tot_nhat):
            tot_nhat = diem
            o_chon = o
    return o_chon

def choi_game():
    ban_co = list(range(1, 10))
    nguoi = input("Bạn chọn X (đi trước) hay O (đi sau)? ").upper()
    ai = O if nguoi == X else X
    luot_hien_tai = X # Mặc định X luôn đi trước

    while True:
        in_ban_co(ban_co)

        # Kiểm tra kết quả trước mỗi lượt
        ket_qua = kiem_tra_thang(ban_co)
        if ket_qua:
            print(f"Kết quả: {ket_qua} THẮNG!")
            break
        if not lay_o_trong(ban_co):
            print("Kết quả: HÒA!")
            break

        if luot_hien_tai == nguoi:
            try:
                chon = int(input(f"Lượt bạn ({nguoi}), chọn ô (1-9): "))
                if chon in lay_o_trong(ban_co):
                    ban_co[chon-1] = nguoi
                    luot_hien_tai = ai
                else:
                    print("Ô đã có người chọn!")
            except:
                print("Vui lòng nhập số!")
        else:
            print("AI đang suy nghĩ...")
            o_ai = tim_nuoc_di_tot_nhat(ban_co, ai)
            ban_co[o_ai-1] = ai
            luot_hien_tai = nguoi

if __name__ == "__main__":
    choi_game()
```

```python
#Bài tập 2 ở lớp

import math

# Khởi tạo bàn cờ 3x3
ban_co = [[' ' for _ in range(3)] for _ in range(3)]

def kiem_tra_thang_thua(b):
    # Kiểm tra hàng và cột
    for i in range(3):
        if b[i][0] == b[i][1] == b[i][2] and b[i][0] != ' ':
            return b[i][0]
        if b[0][i] == b[1][i] == b[2][i] and b[0][i] != ' ':
            return b[0][i]

    # Kiểm tra đường chéo
    if b[0][0] == b[1][1] == b[2][2] and b[0][0] != ' ':
        return b[0][0]
    if b[0][2] == b[1][1] == b[2][0] and b[0][2] != ' ':
        return b[0][2]

    # Kiểm tra hòa
    for hang in b:
        if ' ' in hang: return None
    return 'Hoa'

def thuat_toan_alpha_beta(b, do_sau, alpha, beta, la_nguoi_toi_uu):
    ket_qua = kiem_tra_thang_thua(b)
    if ket_qua == 'O': return 1
    if ket_qua == 'X': return -1
    if ket_qua == 'Hoa': return 0

    if la_nguoi_toi_uu:
        diem_tot_nhat = -math.inf
        for i in range(3):
            for j in range(3):
                if b[i][j] == ' ':
                    b[i][j] = 'O'
                    diem = thuat_toan_alpha_beta(b, do_sau + 1, alpha, beta, False)
                    b[i][j] = ' '
                    diem_tot_nhat = max(diem, diem_tot_nhat)
                    # Cắt tỉa Alpha
                    alpha = max(alpha, diem)
                    if beta <= alpha:
                        break
        return diem_tot_nhat
    else:
        diem_tot_nhat = math.inf
        for i in range(3):
            for j in range(3):
                if b[i][j] == ' ':
                    b[i][j] = 'X'
                    diem = thuat_toan_alpha_beta(b, do_sau + 1, alpha, beta, True)
                    b[i][j] = ' '
                    diem_tot_nhat = min(diem, diem_tot_nhat)
                    # Cắt tỉa Beta
                    beta = min(beta, diem)
                    if beta <= alpha:
                        break
        return diem_tot_nhat

def nuoc_di_cua_may():
    diem_cao_nhat = -math.inf
    hang_tot_nhat = -1
    cot_tot_nhat = -1

    for i in range(3):
        for j in range(3):
            if ban_co[i][j] == ' ':
                ban_co[i][j] = 'O'
                # Khởi tạo alpha = -vô cùng, beta = +vô cùng
                diem_so = thuat_toan_alpha_beta(ban_co, 0, -math.inf, math.inf, False)
                ban_co[i][j] = ' '
                if diem_so > diem_cao_nhat:
                    diem_cao_nhat = diem_so
                    hang_tot_nhat = i
                    cot_tot_nhat = j

    ban_co[hang_tot_nhat][cot_tot_nhat] = 'O'

def hien_thi_ban_co():
    print("\n-------------")
    for hang in ban_co:
        print(f"| {hang[0]} | {hang[1]} | {hang[2]} |")
        print("-------------")

# --- Chương trình chính ---
print("TIC-TAC-TOE (ALPHA-BETA PRUNING)")
while True:
    hien_thi_ban_co()
    kq = kiem_tra_thang_thua(ban_co)
    if kq:
        print(f"TRÒ CHƠI KẾT THÚC: {kq}")
        break

    # Lượt người (X)
    while True:
        try:
            o_di = int(input("Chọn ô (1-9): ")) - 1
            h, c = divmod(o_di, 3)
            if ban_co[h][c] == ' ':5

                ban_co[h][c] = 'X'
                break
            else: print("Ô đã đánh!")
        except: print("Nhập số 1-9!")

    if kiem_tra_thang_thua(ban_co): continue

    # Lượt máy (O)
    print("Máy đang tính toán...")
    nuoc_di_cua_may()
```
