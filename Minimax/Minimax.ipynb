```python
#Code mẫu
import copy
import math
import random
import numpy as np

X = "X"
O = "O"
EMPTY = None
user = None
ai = None

def initial_state():
    """
    Returns starting state of the board.
    """
    return [[EMPTY, EMPTY, EMPTY],
            [EMPTY, EMPTY, EMPTY],
            [EMPTY, EMPTY, EMPTY]]

def player(board):
    """
    Returns player who has the next turn on a board.
    """
    count = 0
    for i in board:
        for j in i:
            if j:
                count += 1
    if count % 2 != 0:
        return ai
    return user

def actions(board):
    """
    Returns set of all possible actions (i, j) available on the board.
    """
    res = set()
    board_len = len(board)
    for i in range(board_len):
        for j in range(board_len):
            if board[i][j] == EMPTY:
                res.add((i, j))
    return res

def result(board, action):
    """
    Returns the board that results from making move (i, j) on the board.
    """
    curr_player = player(board)
    result_board = copy.deepcopy(board)
    (i, j) = action
    result_board[i][j] = curr_player
    return result_board

def get_horizontal_winner(board):
    # check horizontally
    winner_val = None
    board_len = len(board)
    for i in range(board_len):
        winner_val = board[i][0]
        if winner_val is None:
            continue
        for j in range(board_len):
            if board[i][j] != winner_val:
                winner_val = None
                break
        if winner_val:
            return winner_val
    return winner_val

def get_vertical_winner(board):
    # check vertically
    winner_val = None
    board_len = len(board)
    for i in range(board_len):
        winner_val = board[0][i]
        if winner_val is None:
            continue
        for j in range(board_len):
            if board[j][i] != winner_val:
                winner_val = None
                break
        if winner_val:
            return winner_val
    return winner_val

def get_diagonal_winner(board):
    # check diagonally
    board_len = len(board)

    # Check primary diagonal
    winner_val = board[0][0]
    if winner_val is not None:
        for i in range(board_len):
            if board[i][i] != winner_val:
                winner_val = None
                break
        if winner_val:
            return winner_val

    # Check secondary diagonal
    winner_val = board[0][board_len - 1]
    if winner_val is not None:
        for i in range(board_len):
            j = board_len - 1 - i
            if board[i][j] != winner_val:
                winner_val = None
                break
        if winner_val:
            return winner_val

    return None

def winner(board):
    """
    Returns the winner of the game, if there is one.
    """
    return get_horizontal_winner(board) or get_vertical_winner(board) or get_diagonal_winner(board) or None

def terminal(board):
    """
    Returns True if game is over, False otherwise.
    """
    if winner(board) is not None:
        return True
    for i in board:
        for j in i:
            if j == EMPTY:
                return False
    return True

def utility(board):
    """
    Returns 1 if X has won the game, -1 if O has won, 0 otherwise.
    """
    winner_val = winner(board)
    if winner_val == X:
        return 1
    elif winner_val == O:
        return -1
    return 0

def maxValue(state):
    if terminal(state):
        return utility(state)
    v = -math.inf
    for action in actions(state):
        v = max(v, minValue(result(state, action)))
    return v

def minValue(state):
    if terminal(state):
        return utility(state)
    v = math.inf
    for action in actions(state):
        v = min(v, maxValue(result(state, action)))
    return v

def minimax(board):
    """
    Returns the optimal action for the current player on the board.
    """
    current_player = player(board)
    move = None

    if current_player == X:
        v = -math.inf
        for action in actions(board):
            check = minValue(result(board, action))
            if check > v:
                v = check
                move = action
    else:
        v = math.inf
        for action in actions(board):
            check = maxValue(result(board, action))
            if check < v:
                v = check
                move = action
    return move

if __name__ == "__main__":
    board = initial_state()
    ai_turn = False
    print("Choose a player")
    user = input().upper()
    if user == "X":
        ai = "O"
    else:
        ai = "X"

    while True:
        game_over = terminal(board)
        playr = player(board)

        if game_over:
            winner_val = winner(board)
            if winner_val is None:
                print("Game Over: Tie.")
            else:
                print(f"Game Over: {winner_val} wins.")
            break

        if user != playr and not game_over:
            if ai_turn:
                move = minimax(board)
                board = result(board, move)
                ai_turn = False
                print(np.array(board))
        elif user == playr and not game_over:
            ai_turn = True
            print("Enter the position to move (row, col)")
            i = int(input("Row: "))
            j = int(input("Col: "))
            if board[i][j] == EMPTY:
                board = result(board, (i, j))
                print(np.array(board))
```

```python
#Code

import copy
import math
import numpy as np

# Định nghĩa các hằng số
X = "X"
O = "O"
TRONG = None

def trang_thai_dau():
    """Khởi tạo bàn cờ trống 3x3"""
    return [[TRONG, TRONG, TRONG],
            [TRONG, TRONG, TRONG],
            [TRONG, TRONG, TRONG]]

def nguoi_choi_hien_tai(ban_co):
    """Xác định ai là người đi lượt tiếp theo"""
    so_luong_x = sum(hang.count(X) for hang in ban_co)
    so_luong_o = sum(hang.count(O) for hang in ban_co)
    return O if so_luong_x > so_luong_o else X

def cac_nuoc_di_hop_le(ban_co):
    """Trả về tập hợp các tọa độ (dong, cot) còn trống"""
    return {(r, c) for r in range(3) for c in range(3) if ban_co[r][c] == TRONG}

def ket_qua_nuoc_di(ban_co, nuoc_di):
    """Tạo ra một bản sao bàn cờ mới sau khi thực hiện nước đi"""
    dong, cot = nuoc_di
    if ban_co[dong][cot] is not TRONG:
        raise Exception("Nước đi không hợp lệ!")

    ban_co_moi = copy.deepcopy(ban_co)
    ban_co_moi[dong][cot] = nguoi_choi_hien_tai(ban_co)
    return ban_co_moi

def kiem_tra_nguoi_thang(ban_co):
    """Kiểm tra xem ai đã thắng (X, O hoặc None)"""
    # Kiểm tra hàng và cột
    for i in range(3):
        if ban_co[i][0] == ban_co[i][1] == ban_co[i][2] != TRONG:
            return ban_co[i][0]
        if ban_co[0][i] == ban_co[1][i] == ban_co[2][i] != TRONG:
            return ban_co[0][i]

    # Kiểm tra đường chéo
    if ban_co[0][0] == ban_co[1][1] == ban_co[2][2] != TRONG:
        return ban_co[0][0]
    if ban_co[0][2] == ban_co[1][1] == ban_co[2][0] != TRONG:
        return ban_co[0][2]

    return None

def tro_choi_ket_thuc(ban_co):
    """Trả về True nếu game kết thúc, False nếu còn chơi được"""
    if kiem_tra_nguoi_thang(ban_co) is not None:
        return True
    return all(o != TRONG for hang in ban_co for o in hang)

def gia_tri_diem(ban_co):
    """Hàm lượng giá: X thắng = 1, O thắng = -1, Hòa = 0"""
    nguoi_thang = kiem_tra_nguoi_thang(ban_co)
    if nguoi_thang == X: return 1
    if nguoi_thang == O: return -1
    return 0

# --- THUẬT TOÁN MINIMAX VỚI ALPHA-BETA PRUNING ---

def max_value(ban_co, alpha, beta):
    if tro_choi_ket_thuc(ban_co):
        return gia_tri_diem(ban_co)
    v = -math.inf
    for nuoc_di in cac_nuoc_di_hop_le(ban_co):
        v = max(v, min_value(ket_qua_nuoc_di(ban_co, nuoc_di), alpha, beta))
        alpha = max(alpha, v)
        if beta <= alpha: break # Cắt tỉa
    return v

def min_value(ban_co, alpha, beta):
    if tro_choi_ket_thuc(ban_co):
        return gia_tri_diem(ban_co)
    v = math.inf
    for nuoc_di in cac_nuoc_di_hop_le(ban_co):
        v = min(v, max_value(ket_qua_nuoc_di(ban_co, nuoc_di), alpha, beta))
        beta = min(beta, v)
        if beta <= alpha: break # Cắt tỉa
    return v

def minimax_ai(ban_co):
    """Tìm nước đi tối ưu nhất cho AI"""
    nguoi_hien_tai = nguoi_choi_hien_tai(ban_co)
    cac_nuoc = cac_nuoc_di_hop_le(ban_co)

    if nguoi_hien_tai == X:
        gia_tri_tot_nhat = -math.inf
        nuoc_di_tot_nhat = None
        for nuoc_di in cac_nuoc:
            v = min_value(ket_qua_nuoc_di(ban_co, nuoc_di), -math.inf, math.inf)
            if v > gia_tri_tot_nhat:
                gia_tri_tot_nhat = v
                nuoc_di_tot_nhat = nuoc_di
    else:
        gia_tri_tot_nhat = math.inf
        nuoc_di_tot_nhat = None
        for nuoc_di in cac_nuoc:
            v = max_value(ket_qua_nuoc_di(ban_co, nuoc_di), -math.inf, math.inf)
            if v < gia_tri_tot_nhat:
                gia_tri_tot_nhat = v
                nuoc_di_tot_nhat = nuoc_di

    return nuoc_di_tot_nhat

# --- CHƯƠNG TRÌNH CHÍNH ---
if __name__ == "__main__":
    ban_co = trang_thai_dau()
    print("Chào mừng đến với AI Tic-Tac-Toe!")
    ban_chon = input("Bạn muốn chọn X hay O? ").upper()

    while not tro_choi_ket_thuc(ban_co):
        print("\nBàn cờ hiện tại:")
        print(np.array(ban_co))

        nguoi_di = nguoi_choi_hien_tai(ban_co)

        if nguoi_di == ban_chon:
            print(f"Lượt của bạn ({ban_chon}):")
            r = int(input("Nhập dòng (0-2): "))
            c = int(input("Nhập cột (0-2): "))
            if (r, c) in cac_nuoc_di_hop_le(ban_co):
                ban_co = ket_qua_nuoc_di(ban_co, (r, c))
            else:
                print("Vị trí đã có người đi hoặc không hợp lệ!")
        else:
            print("AI đang tính toán...")
            nuoc_ai = minimax_ai(ban_co)
            ban_co = ket_qua_nuoc_di(ban_co, nuoc_ai)

    print("\nKẾT QUẢ CUỐI CÙNG:")
    print(np.array(ban_co))
    winner = kiem_tra_nguoi_thang(ban_co)
    if winner: print(f"Người thắng là: {winner}")
    else: print("Hòa!")
```

```python

#Bài 1 ở lơp
import math

# Khởi tạo bàn cờ 3x3 trống
ban_co = [[' ' for _ in range(3)] for _ in range(3)]

def kiem_tra_thang_thua(b):
    # Kiểm tra các hàng
    for hang in range(3):
        if b[hang][0] == b[hang][1] == b[hang][2] and b[hang][0] != ' ':
            return b[hang][0]

    # Kiểm tra các cột
    for cot in range(3):
        if b[0][cot] == b[1][cot] == b[2][cot] and b[0][cot] != ' ':
            return b[0][cot]

    # Kiểm tra đường chéo
    if b[0][0] == b[1][1] == b[2][2] and b[0][0] != ' ':
        return b[0][0]
    if b[0][2] == b[1][1] == b[2][0] and b[0][2] != ' ':
        return b[0][2]

    # Kiểm tra xem còn ô trống không
    o_trong = False
    for hang in b:
        if ' ' in hang:
            o_trong = True

    if not o_trong:
        return 'Hoa' # Kết quả hòa

    return None # Trò chơi tiếp tục

def thuat_toan_minimax(b, do_sau, la_nguoi_toi_uu):
    ket_qua = kiem_tra_thang_thua(b)

    # Điểm số: Máy (O) thắng: 1, Người (X) thắng: -1, Hòa: 0
    if ket_qua == 'O': return 1
    if ket_qua == 'X': return -1
    if ket_qua == 'Hoa': return 0

    if la_nguoi_toi_uu:
        diem_tot_nhat = -math.inf
        for i in range(3):
            for j in range(3):
                if b[i][j] == ' ':
                    b[i][j] = 'O'
                    diem = thuat_toan_minimax(b, do_sau + 1, False)
                    b[i][j] = ' '
                    diem_tot_nhat = max(diem, diem_tot_nhat)
        return diem_tot_nhat
    else:
        diem_tot_nhat = math.inf
        for i in range(3):
            for j in range(3):
                if b[i][j] == ' ':
                    b[i][j] = 'X'
                    diem = thuat_toan_minimax(b, do_sau + 1, True)
                    b[i][j] = ' '
                    diem_tot_nhat = min(diem, diem_tot_nhat)
        return diem_tot_nhat

def nuoc_di_cua_may():
    diem_cao_nhat = -math.inf
    hang_tot_nhat = -1
    cot_tot_nhat = -1

    for i in range(3):
        for j in range(3):
            if ban_co[i][j] == ' ':
                ban_co[i][j] = 'O'
                diem_so = thuat_toan_minimax(ban_co, 0, False)
                ban_co[i][j] = ' '
                if diem_so > diem_cao_nhat:
                    diem_cao_nhat = diem_so
                    hang_tot_nhat = i
                    cot_tot_nhat = j

    ban_co[hang_tot_nhat][cot_tot_nhat] = 'O'

def hien_thi_ban_co():
    for i, hang in enumerate(ban_co):
        print(" | ".join(hang))
        if i < 2:
            print("-" * 9)

# --- Vòng lặp trò chơi chính ---
print("Chào mừng bạn đến với Tic-Tac-Toe! Bạn là X, Máy là O.")
while True:
    hien_thi_ban_co()
    ket_qua = kiem_tra_thang_thua(ban_co)
    if ket_qua:
        if ket_qua == 'Hoa': print("Kết quả: Hòa!")
        else: print(f"Kết quả: {ket_qua} thắng!")
        break

    # Lượt người chơi
    while True:
        try:
            vi_tri = int(input("Nhập vị trí (1-9): ")) - 1
            h, c = divmod(vi_tri, 3)
            if ban_co[h][c] == ' ':
                ban_co[h][c] = 'X'
                break
            else:
                print("Ô này đã có người đi!")
        except:
            print("Vui lòng nhập số từ 1 đến 9.")

    if kiem_tra_thang_thua(ban_co): continue

    # Lượt máy
    print("\nMáy đang suy nghĩ...")
    nuoc_di_cua_may()
```
